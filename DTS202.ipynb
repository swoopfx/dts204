```python
def show_boxplot_of_numeric_columns(df):
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        plt.figure(figsize=(8, 4))
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot of {col}')
        plt.show()
```

```python
df["absences"][df["absences"] > 22].count() #used to count the number of students with more than 22 absences
```

```python
df["absences"].value_counts().sort_index(ascending=False) #used to calculate the percentage of students with each number of absences
```

```python
# cap absences at 22 across relevant DataFrames
for _df in (df, df3, df_merged):
    if 'absences' in _df.columns:
        _df['absences'] = _df['absences'].clip(upper=22)

# quick check
print("max absences:", df['absences'].max())
print("count of absences == 22:", df['absences'].eq(22).sum())
```

```python
df3['absences'] = df3['absences'].clip(upper=22)
df3.value_counts('absences').sort_index(ascending=False)
```

```python
def show_barplot_of_categorical_columns(df):
    categorical_cols = df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        plt.figure(figsize=(8, 4))
        sns.countplot(x=df[col], order=df[col].value_counts().index, hue=df['is_pass'])
        plt.title(f'Bar Plot of {col}')
        plt.xticks(rotation=45)
        plt.show()
```

```python
show_barplot_of_categorical_columns(df1)
# check if each category has a balanced data for analysis, if not, we can use techniques like oversampling or undersampling to balance the data for better analysis. 
```

```python
show_barplot_of_categorical_columns(df2)
```

```python
# Categorize 'Grade Total' into performance groups
def categorize_grade_total(grade):
    if grade < 25:
        return 'Low'
    elif grade < 35:
        return 'Medium'
    else:
        return 'High'

df3['performance_group'] = df3['Grade Total'].apply(categorize_grade_total)
df3['performance_group'].value_counts()
```
