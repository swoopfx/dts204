```python
# Categorize 'Grade Total' into performance groups
def categorize_grade_total(grade):
    if grade < 25:
        return 'Low'
    elif grade < 35:
        return 'Medium'
    else:
        return 'High'

df3['performance_group'] = df3['Grade Total'].apply(categorize_grade_total)
df3['performance_group'].value_counts()
```

```python
df_merged = pd.concat([df1, df2, df3], axis=1) #used to merge the three dataframes into one dataframe
df_merged.head()
```

```python
# 1. Create composite feature: 'studytime' and 'failures'
df_merged['study_failure_composite'] = df_merged['studytime'] / (df_merged['failures'] + 1)

# 2. Convert absence counts into 'Regular' or 'Irregular' attendance categories
def categorize_attendance(absences):
    if absences < 10:
        return 'Regular'
    else:
        return 'Irregular'

if 'absences' in df_merged.columns:
    df_merged['attendance_category'] = df_merged['absences'].apply(categorize_attendance)
    df_merged.drop('absences', axis=1, inplace=True)

# # Encode attendance_category to numeric (Regular -> 1, Irregular -> 0 or vice versa)
# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()
# df_merged['attendance_category'] = le.fit_transform(df_merged['attendance_category'])

# df_merged[['studytime', 'failures', 'study_failure_composite', 'attendance_category']].head()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Identify categorical and numerical columns in df_merged
categorical_cols = df_merged.select_dtypes(include=['object']).columns
numerical_cols = df_merged.select_dtypes(include=['int64', 'float64']).columns

# Initialize the LabelEncoder and StandardScaler
le = LabelEncoder()
scaler = StandardScaler()

# 1. Encode categorical variables using Label Encoding
for col in categorical_cols:
    df_merged[col] = le.fit_transform(df_merged[col].astype(str))

# 2. Scale numerical features using Standard Scaling
# Excluding 'is_pass' if it exists, as it is the target variable
features_to_scale = [col for col in numerical_cols if col != 'is_pass']
df_merged[features_to_scale] = scaler.fit_transform(df_merged[features_to_scale])

# Ensure the final dataset is fully numeric and ready for modelling
print(df_merged.info())
df_merged.head()
```

```python
#Remove the G1, G2, G3 columns
df_merged.drop(['G1', 'G2', 'G3'], axis=1, inplace=True)
df_merged.head()
```

```python
from sklearn.model_selection import train_test_split

df_merged = df_merged.groupby(level=0).first()
df_merged.head()
# Define features (X) and target (y)
# We use 'is_pass' as the target variable
X = df_merged.drop('is_pass', axis=1)
y = df_merged['is_pass']


# Split the dataset into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training set dimensions: X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"Testing set dimensions: X_test: {X_test.shape}, y_test: {y_test.shape}")
```

```python
y.head()
```

```python
import time
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd

# Initialize models
models = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42)
}

results = []

for name, model in models.items():
    print(f"--- Training and Evaluating {name} ---")
    
    # Measure computational complexity (training time)
    start_time = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start_time
    
    # Predict on testing set
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)
    
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Training Time: {train_time:.4f} seconds\n")
    print("Confusion Matrix:")
    print(conf_matrix)
    print("\n Classification Report:")
    print(class_report)
    print("="*50 + "\n")
    
    # Store results for comparison
    results.append({
        'Model': name,
        'Accuracy': accuracy,
        'Training Time (s)': train_time
    })

# Compare models
comparison_df = pd.DataFrame(results)
print("--- Model Comparison Summary ---")
print(comparison_df.to_string(index=False))


```
